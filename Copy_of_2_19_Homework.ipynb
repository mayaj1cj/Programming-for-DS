```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import cv2
```

```python


from skimage import io

photo = io.imread('numpy_nature')

print(photo)

photo[0,1] = (21, 0, 0)

# plt.imshow(photo[0,1])

# red_photo

# photo.shape = (200, 300, 3)
```

```python


def red_filter(numpy_nature):
    filtered_photo = np.copy(numpy_nature)
    filtered_photo[:,:,1:] = 0
    return filtered_photo

def green_filter(numpy_nature):
    filtered_photo = np.copy(numpy_nature)
    filtered_photo[:,:,[0,2]] = 0
    return filtered_photo

def blue_filter(numpy_nature):
    filtered_photo = np.copy(numpy_nature)
    filtered_photo[:,:,0:2] = 0
    return filtered_photo


image_path = 'numpy_nature.jpg'
photo = cv2.imread(image_path)


photo = cv2.cvtColor(photo, cv2.COLOR_BGR2RGB)


red_filtered_photo = red_filter(photo)
green_filtered_photo = green_filter(photo)
blue_filtered_photo = blue_filter(photo)


plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(red_filtered_photo)
plt.title('Red Filtered Photo')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(green_filtered_photo)
plt.title('Green Filtered Photo')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(blue_filtered_photo)
plt.title('Blue Filtered Photo')
plt.axis('off')

plt.show()

```
